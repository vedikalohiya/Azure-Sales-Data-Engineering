## Doing Transformation for all the tables


```python
# Import required libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Stop any existing Spark session (safe restart)
try:
    spark.stop()
except:
    pass

# Create Spark session
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Sales Data Analysis") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()

# Load dataset (USE YOUR FULL PATH)
df = spark.read.csv(
    "D:/Projects/Azure-Sales-Data-Engineering-MyVersion/archive/olist_order_items_dataset.csv",
    header=True,
    inferSchema=True
)

# Create new column (your transformation ⭐)
df = df.withColumn("revenue", col("price") + col("freight_value"))

# Show output
df.show(5)

# Optional: print schema
df.printSchema()
```
